```python
#    

#   
exclude = set(string.punctuation)
#      
morpher = MorphAnalyzer()

#   -
sw = get_stop_words("ru")
```

```python
#     
def preprocess_txt(line):
    spls = "".join(i for i in str(line).strip() if i not in exclude).split()
    spls = [morpher.parse(i.lower())[0].normal_form for i in spls]
    spls = [i for i in spls if i not in sw and i != ""]
    return spls

df['text'] = df['name'].apply(lambda x: preprocess_txt(str(x)))
```

```python
#   fasttext  
sentences = [i for i in df['text'] if len(i) > 2]
modelFT = FastText(sentences=sentences, vector_size=20, min_count=1, window=5)
```

```python
#  ,          ANN 
#   
ft_index_shop = annoy.AnnoyIndex(100 ,'angular')
index_map_shop = {}
counter = 0

for i in tqdm_notebook(range(len(df))):
    n_ft = 0
    index_map_shop[counter] = (df.iloc[i, 3], df.iloc[i, 2])
    vector_ft = np.zeros(100)
    #     
    for word in df.text:
        if word in modelFT.wv.index_to_key:
            vector_ft += modelFT.wv[word]
            n_ft += 1
    if n_ft > 0:
        vector_ft = vector_ft / n_ft
    ft_index_shop.add_item(counter, vector_ft)
    counter += 1

# 
ft_index_shop.build(10)
```
